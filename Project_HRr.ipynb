### Bien sûr ! Voici une explication des différentes colonnes dans le jeu de données :

1. **Age**: L'âge de l'employé.
2. **Attrition**: Indique si l'employé a quitté l'entreprise (Yes) ou non (No).
3. **BusinessTravel**: Fréquence des déplacements professionnels (Travel_Rarely, Travel_Frequently, Non-Travel).
4. **DailyRate**: Le salaire quotidien de l'employé.
5. **Department**: Le département dans lequel travaille l'employé (par exemple, Sales, Research & Development, etc.).
6. **DistanceFromHome**: La distance entre le lieu de travail et le domicile de l'employé.
7. **Education**: Niveau d'éducation de l'employé (1 = Below College, 2 = College, 3 = Bachelor, 4 = Master, 5 = Doctor).
8. **EducationField**: Domaine d'études de l'employé (par exemple, Life Sciences, Medical, etc.).
9. **EmployeeCount**: Nombre d'employés (constamment égal à 1 dans ce jeu de données).
10. **EmployeeNumber**: Numéro d'identification unique de l'employé.
11. **EnvironmentSatisfaction**: Satisfaction de l'employé vis-à-vis de l'environnement de travail (1 = Low, 2 = Medium, 3 = High, 4 = Very High).
12. **Gender**: Genre de l'employé (Male, Female).
13. **HourlyRate**: Taux horaire de l'employé.
14. **JobInvolvement**: Niveau d'implication dans le travail (1 = Low, 2 = Medium, 3 = High, 4 = Very High).
15. **JobLevel**: Niveau hiérarchique de l'employé (1 = Entry Level, 2 = Intermediate, 3 = Senior, 4 = Manager, 5 = Director).
16. **JobRole**: Poste occupé par l'employé (par exemple, Sales Executive, Research Scientist, etc.).
17. **JobSatisfaction**: Satisfaction de l'employé vis-à-vis de son poste (1 = Low, 2 = Medium, 3 = High, 4 = Very High).
18. **MaritalStatus**: État civil de l'employé (Single, Married, Divorced).
19. **MonthlyIncome**: Revenu mensuel de l'employé.
20. **MonthlyRate**: Taux mensuel de l'employé.
21. **NumCompaniesWorked**: Nombre d'entreprises pour lesquelles l'employé a travaillé précédemment.
22. **Over18**: Indique si l'employé est majeur (Y) ou non (N).
23. **OverTime**: Indique si l'employé effectue des heures supplémentaires (Yes) ou non (No).
24. **PercentSalaryHike**: Augmentation salariale en pourcentage.
25. **PerformanceRating**: Évaluation de la performance de l'employé (1 = Low, 2 = Good, 3 = Excellent, 4 = Outstanding).
26. **RelationshipSatisfaction**: Satisfaction de l'employé vis-à-vis de ses relations professionnelles (1 = Low, 2 = Medium, 3 = High, 4 = Very High).
27. **StandardHours**: Nombre d'heures standard de travail (constamment égal à 80 dans ce jeu de données).
28. **StockOptionLevel**: Niveau d'options d'achat d'actions de l'employé (0 = None, 1 = Low, 2 = Medium, 3 = High).
29. **TotalWorkingYears**: Nombre total d'années de travail de l'employé.
30. **TrainingTimesLastYear**: Nombre de formations suivies par l'employé l'année dernière.
31. **WorkLifeBalance**: Équilibre entre vie professionnelle et vie personnelle de l'employé (1 = Bad, 2 = Good, 3 = Better, 4 = Best).
32. **YearsAtCompany**: Nombre d'années passées par l'employé dans l'entreprise.
33. **YearsInCurrentRole**: Nombre d'années passées par l'employé dans son poste actuel.
34. **YearsSinceLastPromotion**: Nombre d'années depuis la dernière promotion de l'employé.
35. **YearsWithCurrManager**: Nombre d'années passées par l'employé avec son manager actuel.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn import preprocessing
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix, roc_auc_score

def print_score(clf, X_train, X_test, y_train, y_test, train=True):
    lb = preprocessing.LabelBinarizer()
    lb.fit(y_train)
    if train:  
        res = clf.predict(X_train)
        print("Train Result:\n")
        print("accuracy score: {0:.4f}\n".format(accuracy_score(y_train, res)))
        print("Classification Report: \n{}\n".format(classification_report(y_train, res)))
        print("Confusion Matrix: \n{}\n".format(confusion_matrix(y_train, res)))
        print("ROC AUC: {0:.4f}\n".format(roc_auc_score(lb.transform(y_train), lb.transform(res))))
    elif train==False :
        res_test = clf.predict(X_test)
        print("Train Result:\n")
        print("accuracy score: {0:.4f}\n".format(accuracy_score(y_test, res_test)))
        print("Classification Report: \n{}\n".format(classification_report(y_test, res_test)))
        print("Confusion Matrix: \n{}\n".format(confusion_matrix(y_test, res_test)))
        print("ROC AUC: {0:.4f}\n".format(roc_auc_score(lb.transform(y_test), lb.transform(res_test))))
```

```python
print_score(clf, X_train, X_test, y_train, y_test,train=True)
print_score(clf, X_train, X_test, y_train, y_test,train=False)
```
